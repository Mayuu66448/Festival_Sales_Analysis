# Extract and cleaning of the data

```python
# we will delete null columns

df.drop(['Status','unnamed1'] , axis = 1 , inplace = True)
df.info()
```

```python
# now we will delete null value from table 

df.dropna(inplace = True)
```

# EXPLORATORY DATA  ANALYSIS

# Gender

```python
# plotting a bar chart for gender and its count
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title("Count orders by Gender", fontsize = 14)
```

```python
# plotting bar chart for total amount by Gender wise

ab = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set({'figure.figsize' : (7,2)})
ab = sns.barplot(x='Gender',y = 'Amount', data= ab)
for bars in ab.containers:
    ab.bar_label(bars, fmt= '%.0f')
    
plt.title("Total Amount by Gender", fontsize = 14)
```

 from above graphs we can see most of the buyers are Female and most money sepnt by Female.

# Age

```python
#Plotting bar chart for age group and its count
sns.set({'figure.figsize' : (10,5)})
a = sns.countplot(data=df , x='Age Group', hue = 'Gender')
for bars in a.containers:
    a.bar_label(bars)
plt.title("Count orders by age group", fontsize = 14)
```

```python
# Plotting bar chart for age broup and its total Amount 

age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending=True)

ab= sns.barplot(data=age, x='Age Group', y = 'Amount')
for bars in ab.containers:
    ab.bar_label(bars, fmt = '%.0f' )
plt.title("Total Amount by Age Group", fontsize = 14)
```

 as we can see most of the buyers are from age group 26-35 yrs female.

# State

```python
#Plotting Bar chart for State and its number of orders

order= df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set({'figure.figsize':(15,5)})
order= sns.barplot(data=order, x= 'State', y = 'Orders')

for bars in order.containers:
    order.bar_label(bars)
plt.title(" Top 10 State BY Orders", fontsize = 14)
```

```python
#Plotting bar chart for State and its total amount

amount = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by = "Amount" , ascending = False).head(10)
sns.set({'figure.figsize' : (15,5)})
amount = sns.barplot(data= amount, x= 'State', y = 'Amount')

for bars in amount.containers:
    amount.bar_label(bars , fmt =  '%d')
plt.title("Top 10 State by Total Amount", fontsize = 14)
                                                                                                                    
```

as we can see total order and Amount/Sales is from Utar Pradesh, maharashtra and Kerala Respectively.

# Marital_Status

```python
# Plotting Bar chart for Marital_Status and its count.

sns.set({'figure.figsize': (5,3)})
ax = sns.countplot( data = df, x='Marital_Status' )
for bars in ax.containers:
    ax.bar_label(bars )
plt.title("Count orders by Marital_Status", fontsize = 14)
```

```python
# Plotting Marital_Status and its Total Amount

ax = df.groupby(['Marital_Status' , 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set({'figure.figsize' : (8,3)})
ax = sns.barplot(data = ax , x= 'Marital_Status', y = 'Amount', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars, fmt = '%d')
    
plt.title("Total Amount by Marital_Status", fontsize = 14)
```

as we can see  most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
# Plotting Occupation and its count

sns.set({'figure.figsize' : (20,5)})

ax = sns.countplot( data=df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title("Count orders by Occupation", fontsize = 14)
```

```python
# Plotting Occupation and It's Total Amount

ax = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(18,5)})

ax = sns.barplot(data = ax, x='Occupation', y='Amount')
for bars in ax.containers:
    ax.bar_label(bars, fmt = '%d')
    
plt.title("Total Amount by Occupation", fontsize = 14)
```

as wee can see most of buyers are working in It sector, Healthcare and Aviation respectively.

# Product_Category

```python
# Plotting Top 10 Product_Category and its Count

sns.set({'figure.figsize':(24,10)})
ab =df.groupby(['Product_Category'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
ab = sns.barplot(data = ab, x = 'Product_Category', y = 'Orders')
                                                                                
for bars in ab.containers:
    ab.bar_label(bars)
    
plt.title('Top 10 Product_Category by orders', fontsize = 14)
```

```python
# Plotting Top 10 Product_Category and its Amount

ax = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set({'figure.figsize':(20,5)})
ab = sns.barplot(data = ax , x = 'Product_Category', y = 'Amount' )

for bars in ab.containers:
    ab.bar_label(bars, fmt = '%d')

plt.title("Top 10 Product_Category by Amount", fontsize = 14)
```

as we can see most of the sold products are from Food, Clothing and Electronics category
